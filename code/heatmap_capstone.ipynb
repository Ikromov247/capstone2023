```python
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(111, projection='3d')

ax.set_ylabel("NH3")
ax.set_xlabel("H2S")
ax.set_zlabel("TVOCs")
ax.scatter( x["H2S"], x["NH3"],x["TVOCs"] , c=x["ou"], cmap='magma', linewidth=0.5, s=1)
```
