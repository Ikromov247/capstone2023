#Features to scale:
OU, Temperature, Humidity, Windspeed.


```python
# Scale
def feature_scale(col):
  # std normalization
  sigma = col.std()
  mean = col.mean()
  return (col - mean) / sigma
```

```python
def min_max_scale(col):
  # minmax normalization
  range = col.max() - col.min()
  mean = col.mean()
  return (col - mean) / range
```

```python
# STD NORMALIZED VALUES HERE

# Store normalized feature values as arr in dict with names as keys

to_scale = ["ou", "temperature", "humidity", "windspeed"]
std_scaled = {}
for feature in to_scale:
  temp = feature_scale(df[feature])
  range = temp.max() - temp.min()
  print(feature, "range:", range)
  std_scaled[feature] = temp

# ttemp = pd.DataFrame(std_scaled)
```

```python
# MINMAX NORMALIZED VALUES HERE

to_scale = ["ou", "temperature", "humidity", "windspeed"]
minmax_scaled = {}
for feature in to_scale:
  temp = feature_scale(df[feature])
  range = temp.max() - temp.min()
  print(feature, "range:", range)
  minmax_scaled[feature] = temp

# ttemp = pd.DataFrame(minmax_scaled)
```

```python
# WIND DIRECTION NAMES
original_values = df.winddirection.unique()
translated = ['South', 'South-South-East', 'North', 'North-West', 'West-South-West', 'West', 'West-North-West', 'South-South-West', 'North-North-East', 'North-North-West',
        'East-Southeast', 'Southwest', 'Northeast', 'Southeast', 'East', 'East-Northeast', '-']
```

```python
for v, x in zip(original_values, translated):
  print(f"{v}-->{x}")
```

```python
# dict with direction name and index as key-value pair
label_int = {}
i=0
for v in original_values:
  label_int[v] = i
  i+=1
```

```python
# INPLACE IS FALSE. TRUE ONLY IF SURE ABOUT REPLACING
df.winddirection.replace(label_int, inplace=False)
```

```python
# PREPARE AND SAVE STD NORMALIZED DF
std_temp = df

for col in std_temp.columns:
  if col in to_scale:
    std_temp[col] = std_scaled[col]

std_temp.winddirection.replace(label_int, inplace=True)
std_temp.head(5)
```
