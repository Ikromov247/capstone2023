#Data loading

```python
data_path = "/content/drive/MyDrive/capstone_data/2022-12-01-2023-07-01(109┴÷┴í).xls.xlsx"
dataset = pd.read_excel(data_path)

dataset.head(10)
```

```python
x_train = dataset.drop(["ou","pozip"], axis=1, inplace=False).iloc[5:, 1:4]
y_train = dataset.iloc[5:, 1]
```

```python
x_test = dataset.drop(["ou","pozip"], axis=1, inplace=False).iloc[:5, 1:4]
y_test = dataset.iloc[:5, 1]
```

```python
# Fitting Random Forest Regression to the dataset
# import the regressor
from sklearn.ensemble import RandomForestRegressor

# create regressor object
regressor = RandomForestRegressor(n_estimators=1000,
								random_state=0)

# fit the regressor with x and y data
```

```python
regressor.fit(x_train, y_train)
```

```python
regressor.predict(x_test.iloc[:,:])
```
